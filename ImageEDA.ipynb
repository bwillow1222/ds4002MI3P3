```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# ANALYZE DATA FOLDER

# Specify the main folder containing classified subfolders
main_folder = '/Users/adamsnyder/Documents/DS 4002/pet_images'

# Dictionary to store the count of images in each classification folder
classification_counts = {}

# Loop over each subfolder in the main folder
for classification in os.listdir(main_folder):
    classification_folder = os.path.join(main_folder, classification)
    
    # Check if it's a directory (i.e., a classification folder)
    if os.path.isdir(classification_folder):
        # Count the number of image files in the classification folder
        count = len([file for file in os.listdir(classification_folder)
                     if file.lower().endswith(('.jpg', '.jpeg', '.png'))])  # Adjust as needed
        classification_counts[classification] = count
```

```python
# CREATE FREQUENCY CHART

# Plotting the frequency distribution
classifications = list(classification_counts.keys())
counts = list(classification_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(classifications, counts, color='skyblue')
plt.xlabel('Breed')
plt.ylabel('Number of Images')
plt.title('Frequency Distribution of Classified Images')
plt.xticks(rotation=45, ha="right")  # Rotate labels for readability if needed
plt.tight_layout()
plt.show()
```

```python
# CREATE TABLE OF COUNTS

# Convert the dictionary to a pandas DataFrame for a table format
df = pd.DataFrame(list(classification_counts.items()), columns=['Classification', 'Count'])

# Display the table
print(df)
```
